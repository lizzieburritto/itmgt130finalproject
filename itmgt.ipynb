```python
# Setup MongoDB Client
import pymongo

# Replace <database_name> with the name of your target DB
database_name = "db"

# Set up the connection string
connection_string = f"mongodb://localhost:27017"

# Connect to the MongoDB server
client = pymongo.MongoClient(connection_string)

# Access the database
db = client[database_name]
```

```python
# Access collections
patient_db = db["patients"]
appointments_db = db["appointments"]
dentist_db = db["dentists"]

def get_patient(name):
    patient = patient_db.find_one({"name": name})
    return patient

def add_patient(patient):
    patient_db.insert_one(patient)

def add_appointment(appointment):
    result = appointments_db.insert_one(appointment)
    
def get_dentist_appointments(dentist_code):
    appointments = appointments_db.find({"dentist_code": dentist_code})
    return appointments
```
